```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_context('talk',font_scale=1)

data = pd.read_json("optimisation.json")
#data.head()
```

```python
data = data.iloc[20:]
```

```python
def corr_Date(date):
    if date.split(" ")[-1] == "hier":
        return ("hier")
    tab = " ".join(date.split(" ")[-2:])
    return tab

data["Date de publication"] = data["Date de publication"].apply(corr_Date)
#data.head()
```

```python
data.fillna(value=" ",inplace=True)

def list_to_str_poste(liste):
    return(' '.join(liste))

data["Intitulé du poste"] = data["Intitulé du poste"].apply(list_to_str_poste)
#data.head()
```

```python
#data["Intitulé du poste"].value_counts()
```

```python
def list_to_str_competance_et_lieu(liste):
    return(','.join(liste))

def CorrectionTab(string):
    # return(''.join(string.split("\n")))
    string = string.replace("\n","")
    return string

data["competences"] = data["competences"].apply(list_to_str_competance_et_lieu)

#data.head()
#def le_type(elemt):
#    print(type(elemt))
```

```python
listecol=["Date de publication","Intitulé du poste","competences"]
for col in listecol:
    data[col] = data[col].apply(CorrectionTab)
#data.head()
```

```python
#data["lieu"].value_counts()
#data["Type de poste"].value_counts().head()
```

```python
data_taf=data.copy()
data_taf

def drop_n_type_poste(liste):
    liste = [x for x in liste if (x != "\n")]
    liste = [x for x in liste if (x != " \n")]
    return liste

def retourner_salaire(liste):
    if len(liste) == 1:
        return (np.nan)
    else:
        return (liste[-1])
    
def type_de_contrat(liste):
    return liste[-1]

def nom_societe(liste):
    return liste[0]

data_taf["Salaire"] = data_taf["lieu"].apply(retourner_salaire)
data_taf["Type de poste"] = data_taf["Type de poste"].apply(drop_n_type_poste)
data_taf["Type de contrat"] = data_taf["Type de poste"].apply(type_de_contrat)
#data_taf["Type de contrat"].value_counts()
data_taf["Nom de la société"] = data_taf["Type de poste"].apply(nom_societe)
data_taf.replace("Societe Generale","Société générale",inplace=True)
#data_taf["Nom de la société"].value_counts()
data_taf.drop(columns="Type de poste",inplace=True)
```

# Notre jeu de données est trié

```python
def Correctionlieu(liste):
    # return(''.join(string.split("\n")))
    return liste[0]

data_taf["lieu"] = data_taf["lieu"].apply(Correctionlieu)
data_taf["lieu"] = data_taf["lieu"].apply(CorrectionTab)

def CorrectionSalaire(value):
    if type(value) == str:
        var_temporaire=CorrectionTab(value)
        var_temporaire=var_temporaire.replace("30000 - 50000 €/an","30.000,00 € - 50.000,00 € / an")
        return var_temporaire.replace(" selon profil","")
    else:
        return np.NaN

data_taf["Salaire"] = data_taf["Salaire"].apply(CorrectionSalaire)
data_taf["Type de contrat"] = data_taf["Type de contrat"].apply(CorrectionTab)
data_taf["Nom de la société"] = data_taf["Nom de la société"].apply(CorrectionTab)

pd.set_option('display.max_rows', None)
data_taf
```

```python
sns.set_context('talk',font_scale=5)
def ajoute_une_virgule_a_la_fin(chaine_de_comp):
    return(chaine_de_comp+",")

data_taf["comp"]=data_taf["competences"].apply(ajoute_une_virgule_a_la_fin)

liste_de_competance_totale=("").join(data_taf["comp"].tolist()).split(",")[:-1]
#on doit retirer la dernière valeur

data_taf.drop(columns=['comp'],inplace=True)

liste_de_competance_totale.remove('')
plt.figure(figsize=(15,35))
df_de_comp_tot=pd.DataFrame({"compétences":liste_de_competance_totale})
df_de_comp_tot.head(10)


df_de_comp_tot.replace(to_replace="securité", value="sécurité", inplace=True)
df_de_comp_tot.replace(to_replace=["Data-Scientist - CDI","Data Scientist","DATA SCIENTIST"],value="Data Scientist", inplace=True)


sns.set(style="darkgrid")
sns.countplot(y="compétences", data=df_de_comp_tot,order=df_de_comp_tot['compétences'].value_counts().index,palette="Blues_d")
plt.title("Les compétances les plus recherchées",color="#005668",fontsize=30)
plt.grid()
```

```python
liste_salaire_min = []
liste_salaire_max = []

for i in range (df_pertinant.shape[0]):
    for sal in range (len(df_pertinant["competences"].iloc[i].split(","))):
        liste_salaire_min.append(df_pertinant["Salaire_min"].iloc[i])
        liste_salaire_max.append(df_pertinant["Salaire_max"].iloc[i])
        
#print(len(liste_de_comp))
#print(len(liste_salaire_max))
#print(len(liste_salaire_max))
```

```python
d = {'Compétances': liste_de_comp,'Salaire_min_associé':liste_salaire_min,'Salaire_max_associé':liste_salaire_max}
df_comp = pd.DataFrame(data=d)
df_comp["Compétances"].value_counts()
df_comp.replace("securité","sécurité",inplace=True)
df_comp.replace("réseau","réseaux",inplace=True)

fig = plt.figure(figsize=(30,40))
sns.set_context('talk',font_scale=2)
sns.countplot(y=df_comp["Compétances"],order=df_comp["Compétances"].value_counts().index)
plt.title("Les compétances les plus recherchées parmis les salaires mentionnés",color="#775e40",fontsize=40)
```
